```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt

def load_csv(csv_filename):
    """Load a CSV file and handle exceptions."""
    try:
        return pd.read_csv(csv_filename)
    except FileNotFoundError:
        print(f"Error: CSV file '{csv_filename}' not found.")
        return None
    except Exception as e:
        print(f"Error occurred while loading CSV file: {e}")
        return None

def univariate_analysis(data, column, target_column='TARGET', top_n=10):
    """Perform univariate analysis on a specified column."""
    if column not in data.columns:
        print(f"Column '{column}' not found in the dataset.")
        return
    
    try:
        # Group by the specified column and calculate aggregate statistics
        grouped_data = data.groupby(column)[target_column].agg(['sum', 'count', 'mean']).reset_index()
        grouped_data.columns = [column, 'Defaulters', 'Total', 'Defaulter Rate']
        grouped_data.sort_values(by='Total', ascending=False, inplace=True)
        
        # Display a stacked bar plot and summary statistics
        stack_plot(grouped_data.head(top_n), xtick=column, col2='Defaulters', col3='Total')
        print(f"Top {top_n} {column} values:")
        print(grouped_data.head(top_n))
        print("=" * 50)
        print(f"Lowest {top_n} {column} values:")
        print(grouped_data.tail(top_n))
        
        # Check for low-risk conditions (e.g., high approval rate)
        if 'Defaulter Rate' in grouped_data.columns:
            low_risk_values = grouped_data[grouped_data['Defaulter Rate'] < 0.1][column].values
            if len(low_risk_values) > 0:
                print(f"\nLow-risk {column} values (Defaulter Rate < 10%):")
                print(", ".join(map(str, low_risk_values)))
    
    except Exception as e:
        print(f"Error occurred for column '{column}': {e}")

def stack_plot(data, xtick, col2='TARGET', col3='total'):
    """Create a stacked bar plot."""
    unique_values_count = len(data[xtick].unique())
    figsize = (5, 5) if unique_values_count < 5 else (7, 7) if unique_values_count < 10 else (15, 15)
    
    plt.figure(figsize=figsize)
    ind = np.arange(data.shape[0])
    p1 = plt.bar(ind, data[col3].values)
    p2 = plt.bar(ind, data[col2].values, bottom=data[col3].values)
    
    plt.ylabel('Number of Loans')
    plt.title(f'Loan Approval Statistics by {xtick}')
    plt.xticks(ticks=ind, rotation=90, labels=list(data[xtick].values))
    plt.legend((p1[0], p2[0]), ('Non-Defaulters (0)', 'Defaulters (1)'))
    plt.show()

if __name__ == "__main__":
    # Specify the path to the CSV file
    csv_filename = 'data/application_train.csv'
    
    # Load the CSV file
    train_data = load_csv(csv_filename)
    
    if train_data is None:
        exit(1)  # Exit if CSV file loading failed
    
    # List of columns to analyze
    columns_to_analyze = [
        'NAME_CONTRACT_TYPE', 'CODE_GENDER', 'FLAG_OWN_CAR', 'FLAG_OWN_REALTY', 'CNT_CHILDREN', 
        'NAME_TYPE_SUITE', 'NAME_INCOME_TYPE', 'NAME_EDUCATION_TYPE', 'NAME_FAMILY_STATUS', 
        'NAME_HOUSING_TYPE', 'OCCUPATION_TYPE', 'CNT_FAM_MEMBERS', 'REGION_RATING_CLIENT'
    ]

    # Iterate over each column for univariate analysis
    for column in columns_to_analyze:
        print(f"Analyzing column: {column}")
        univariate_analysis(train_data, column, target_column='TARGET', top_n=10)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def load_csv(csv_filename):
    """Load a CSV file and handle exceptions."""
    try:
        return pd.read_csv(csv_filename)
    except FileNotFoundError:
        print(f"Error: CSV file '{csv_filename}' not found.")
        return None
    except Exception as e:
        print(f"Error occurred while loading CSV file: {e}")
        return None

def calculate_repayment_rate(data, column, target_column='TARGET'):
    """Calculate repayment rate (percentage of non-defaulters) for a specific column."""
    repayment_rates = data.groupby(column)[target_column].mean() * 100
    return repayment_rates

def generate_observations(data, columns_to_analyze, target_column='TARGET'):
    """Generate dynamic observations based on specified columns."""
    observations = []

    for column in columns_to_analyze:
        if column not in data.columns:
            print(f"Column '{column}' not found in the dataset. Skipping...")
            continue
        
        if data[column].dtype == 'object':
            # For categorical columns
            repayment_rates = calculate_repayment_rate(data, column, target_column)
            top_category = repayment_rates.idxmax()
            top_repayment_rate = repayment_rates.max()
            observations.append(f"'{top_category}' for '{column}' has the highest repayment rate ({top_repayment_rate:.1f}%).")

    return observations

def plot_loan_approval_stats(data, column, target_column='TARGET', top_n=10):
    """Plot loan approval statistics for a specified column."""
    if column not in data.columns:
        print(f"Column '{column}' not found in the dataset.")
        return
    
    try:
        # Group by the specified column and calculate aggregate statistics
        grouped_data = data.groupby(column)[target_column].agg(['sum', 'count', 'mean']).reset_index()
        grouped_data.columns = [column, 'Defaulters', 'Total', 'Defaulter Rate']
        grouped_data.sort_values(by='Total', ascending=False, inplace=True)
        
        # Display a bar plot and summary statistics
        plt.figure(figsize=(10, 6))
        sns.barplot(x=column, y='Total', data=grouped_data.head(top_n), palette='viridis')
        plt.title(f'Loan Approval Statistics by {column}')
        plt.xlabel(column)
        plt.ylabel('Number of Loans')
        plt.xticks(rotation=90)
        plt.show()

        print(f"Top {top_n} {column} values:")
        print(grouped_data.head(top_n))
        print("=" * 50)
        print(f"Lowest {top_n} {column} values:")
        print(grouped_data.tail(top_n))
        
        # Check for low-risk conditions (e.g., high approval rate)
        if 'Defaulter Rate' in grouped_data.columns:
            low_risk_values = grouped_data[grouped_data['Defaulter Rate'] < 0.1][column].values
            if len(low_risk_values) > 0:
                print(f"\nLow-risk {column} values (Defaulter Rate < 10%):")
                print(", ".join(map(str, low_risk_values)))
    
    except Exception as e:
        print(f"Error occurred for column '{column}': {e}")

def identify_best_borrower_characteristics(data, columns_to_analyze, target_column='TARGET'):
    """Identify characteristics associated with higher repayment rates (non-defaulters)."""
    best_characteristics = {}
    
    for column in columns_to_analyze:
        if column not in data.columns:
            print(f"Column '{column}' not found in the dataset. Skipping...")
            continue
        
        if data[column].dtype == 'object':
            # For categorical columns
            repayment_rates = calculate_repayment_rate(data, column, target_column)
            best_category = repayment_rates.idxmax()
            best_repayment_rate = repayment_rates.max()
            best_characteristics[column] = {'best_category': best_category, 'repayment_rate': best_repayment_rate}
    
    return best_characteristics

def identify_worst_borrower_characteristics(data, columns_to_analyze, target_column='TARGET'):
    """Identify characteristics associated with lower repayment rates (higher default rates)."""
    worst_characteristics = {}
    
    for column in columns_to_analyze:
        if column not in data.columns:
            print(f"Column '{column}' not found in the dataset. Skipping...")
            continue
        
        if data[column].dtype == 'object':
            # For categorical columns
            repayment_rates = calculate_repayment_rate(data, column, target_column)
            worst_category = repayment_rates.idxmin()  # Find category with the lowest repayment rate
            worst_repayment_rate = repayment_rates.min()
            worst_characteristics[column] = {'worst_category': worst_category, 'repayment_rate': worst_repayment_rate}
    
    return worst_characteristics

if __name__ == "__main__":
    # Specify the path to the CSV file
    csv_filename = 'data/application_train.csv'
    
    # Load the CSV file
    train_data = load_csv(csv_filename)
    
    if train_data is None:
        exit(1)  # Exit if CSV file loading failed
    
    # List of columns to analyze
    columns_to_analyze = [
        'NAME_CONTRACT_TYPE', 'CODE_GENDER', 'FLAG_OWN_CAR', 'FLAG_OWN_REALTY', 'CNT_CHILDREN', 
        'NAME_TYPE_SUITE', 'NAME_INCOME_TYPE', 'NAME_EDUCATION_TYPE', 'NAME_FAMILY_STATUS', 
        'NAME_HOUSING_TYPE', 'OCCUPATION_TYPE', 'CNT_FAM_MEMBERS', 'REGION_RATING_CLIENT'
    ]

    # Analyze each column and generate observations
    for column in columns_to_analyze:
        print(f"Analyzing column: {column}")
        plot_loan_approval_stats(train_data, column, target_column='TARGET', top_n=10)

    # Generate observations based on specified columns
    observations = generate_observations(train_data, columns_to_analyze)

    # Print the generated observations
    print("\nAutomated Observations:")
    for idx, obs in enumerate(observations, start=1):
        print(f"{idx}. {obs}")

    # Identify best borrower characteristics based on repayment rates
    best_characteristics = identify_best_borrower_characteristics(train_data, columns_to_analyze)
    print("\nBest Borrower Characteristics:")
    for column, info in best_characteristics.items():
        print(f"For '{column}', '{info['best_category']}' has the highest repayment rate ({info['repayment_rate']:.1f}%).")

    # Identify worst borrower characteristics based on repayment rates
    worst_characteristics = identify_worst_borrower_characteristics(train_data, columns_to_analyze)
    print("\nWorst Borrower Characteristics:")
    for column, info in worst_characteristics.items():
        print(f"For '{column}', '{info['worst_category']}' has the lowest repayment rate ({info['repayment_rate']:.1f}%).")
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, StandardScaler

def create_new_features(data):
    # Numerical features
    data['CREDIT_INCOME_PERCENT'] = data['AMT_CREDIT'] / data['AMT_INCOME_TOTAL']
    data['ANNUITY_INCOME_PERCENT'] = data['AMT_ANNUITY'] / data['AMT_INCOME_TOTAL']
    data['CREDIT_ANNUITY_PERCENT'] = data['AMT_CREDIT'] / data['AMT_ANNUITY']
    data['FAMILY_CNT_INCOME_PERCENT'] = data['AMT_INCOME_TOTAL'] / data['CNT_FAM_MEMBERS']
    data['CREDIT_TERM'] = data['AMT_ANNUITY'] / data['AMT_CREDIT']
    data['BIRTH_EMPLOYED_PERCENT'] = data['DAYS_EMPLOYED'] / data['DAYS_BIRTH']
    data['CHILDREN_CNT_INCOME_PERCENT'] = data['AMT_INCOME_TOTAL'] / (data['CNT_CHILDREN'] + 1)  # Avoid division by zero

    # Difference between credit amount and goods price
    data['CREDIT_GOODS_DIFF'] = data['AMT_CREDIT'] - data['AMT_GOODS_PRICE']

    # Percentage of employed time to registration time and birth time
    data['EMPLOYED_REGISTRATION_PERCENT'] = data['DAYS_EMPLOYED'] / data['DAYS_REGISTRATION']
    data['BIRTH_REGISTRATION_PERCENT'] = data['DAYS_BIRTH'] / data['DAYS_REGISTRATION']
    data['ID_REGISTRATION_DIFF'] = data['DAYS_ID_PUBLISH'] - data['DAYS_REGISTRATION']

    # Additional features
    data['ANNUITY_LENGTH_EMPLOYED_PERCENT'] = data['CREDIT_TERM'] / data['DAYS_EMPLOYED']
    data['AGE_LOAN_FINISH'] = (data['DAYS_BIRTH'] / -365) + (data['AMT_CREDIT'] / data['AMT_ANNUITY']) * (1 / 12)

    # Ratio of car age to employed days and birth days
    data['CAR_AGE_EMP_PERCENT'] = data['OWN_CAR_AGE'] / data['DAYS_EMPLOYED']
    data['CAR_AGE_BIRTH_PERCENT'] = data['OWN_CAR_AGE'] / data['DAYS_BIRTH']

    # Ratio of phone change days to employed days and birth days
    data['PHONE_CHANGE_EMP_PERCENT'] = data['DAYS_LAST_PHONE_CHANGE'] / data['DAYS_EMPLOYED']
    data['PHONE_CHANGE_BIRTH_PERCENT'] = data['DAYS_LAST_PHONE_CHANGE'] / data['DAYS_BIRTH']

def encode_categorical_features(data):
    # One-hot encode categorical features
    categorical_columns = data.select_dtypes(include=['object']).columns.tolist()
    encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
    encoded_data = pd.DataFrame(encoder.fit_transform(data[categorical_columns]))
    encoded_data.columns = encoder.get_feature_names_out(categorical_columns)
    data = pd.concat([data.drop(columns=categorical_columns), encoded_data], axis=1)
    return data

def drop_unnecessary_columns(data):
    # Drop unnecessary columns
    columns_to_drop = [
        'FLAG_DOCUMENT_' + str(i) for i in range(2, 22)
    ]
    data = data.drop(columns=columns_to_drop, axis=1)
    return data

def feature_engineering(data):
    # Create new features
    create_new_features(data)
    
    # Encode categorical features
    data = encode_categorical_features(data)

    # Drop unnecessary columns
    data = drop_unnecessary_columns(data)

    return data

# Usage example:
# Assuming 'train_data' is your DataFrame containing the training data
train_data_processed = feature_engineering(train_data.copy())
```
